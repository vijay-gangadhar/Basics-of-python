```python
#Data Cleaning
#filling the missing values with mean
df_train["Age"]=df_train["Age"].fillna(df_train["Age"].mean().round(1))
df_train.info()
```

```python
df_test["Age"]=df_test["Age"].fillna(df_test["Age"].mean().round(1))
df_test.info()
```

```python
#removing unneccesary columns
df_train.drop(["PassengerId","Name","Cabin","Ticket"],axis=1,inplace=True)
df_train.info()
```

```python
#removing unneccesary columns
df_test.drop(["PassengerId","Name","Cabin","Ticket"],axis=1,inplace=True)
df_test.info()
```

```python
#dropping null values
df_train.dropna(inplace=True)
df_train.info()
```

```python
#dropping null values
df_test.dropna(inplace=True)
df_test.info()
```

```python
#Data Transformation

# label encoding
from sklearn.preprocessing import LabelEncoder
LE = LabelEncoder()
df_train['Sex'] = LE.fit_transform(df_train['Sex'])
df_train['Embarked'] = LE.fit_transform(df_train['Embarked'])
df_train
```

```python
df_test['Sex'] = LE.fit_transform(df_test['Sex'])
df_test['Embarked'] = LE.fit_transform(df_test['Embarked'])
df_test
```

```python
from sklearn.preprocessing import StandardScaler
SS = StandardScaler()
df_train["Age"] = SS.fit_transform(df_train[["Age"]])
df_train["Fare"] = SS.fit_transform(df_train[["Fare"]])
df_train["SibSp"] = SS.fit_transform(df_train[["SibSp"]])
df_train["Parch"] = SS.fit_transform(df_train[["Parch"]])
df_train.head()
```

```python
df_test["Age"] = SS.fit_transform(df_test[["Age"]])
df_test["Fare"] = SS.fit_transform(df_test[["Fare"]])
df_test["SibSp"] = SS.fit_transform(df_test[["SibSp"]])
df_test["Parch"] = SS.fit_transform(df_test[["Parch"]])
df_test.head()
```

```python
Y_train=df_train["Sex"]
df_train.drop("Sex",axis=1,inplace=True)
df_train
```

```python
Y_test=df_test["Sex"]
df_test.drop("Sex",axis=1,inplace=True)
df_test
```

```python
X_train=df_train.iloc[:,1:7]
```

```python
X_test=df_test.iloc[:,0:6]
```

```python
# step5: Model fitting
from sklearn.linear_model import LogisticRegression
logreg = LogisticRegression()
logreg.fit(X_train,Y_train)
logreg.intercept_
```

```python
logreg.coef_
```

```python
# predictions
Y_train_pred = logreg.predict(X_train)
Y_test_pred = logreg.predict(X_test)
Y_train_pred
```

```python
Y_test_pred
```

```python
# step6: Metrics

from sklearn.metrics import confusion_matrix, accuracy_score, recall_score, precision_score, f1_score

cm = confusion_matrix(Y_test,Y_test_pred)
cm
```

```python

ac = accuracy_score(Y_test,Y_test_pred)
print("Accuracy score:", ac.round(2))
```

```python
rs = recall_score(Y_test,Y_test_pred)
print("sensitivity score:", rs.round(2))
```

```python
TN = cm[0,0]
FP = cm[0,1]
TNR = TN/(TN + FP)
print("specificity score:", TNR.round(2))
```

```python
ps = precision_score(Y_test,Y_test_pred)
print("Precision score:", ps.round(2))
```

```python
f1 = f1_score(Y_test,Y_test_pred)
print("F1 score:", f1.round(2))
```

```python
from sklearn.metrics import roc_curve,roc_auc_score

df_test["Y_proba"] = logreg.predict_proba(X_test)[:,1]
df_test.head()

fpr,tpr,dummy = roc_curve(Y_test,df_test["Y_proba"])
```

```python
import matplotlib.pyplot as plt
plt.scatter(fpr,tpr)
plt.plot(fpr,tpr,color='red')
plt.xlabel("False positive Rate")
plt.ylabel("True positive Rate")
plt.show()

print("AUC score:", roc_auc_score(Y_test,df_test["Y_proba"]).round(3))
```
